```python
from sklearn.datasets import load_breast_cancer
from synthcity.plugins import Plugins
from synthcity.metrics import evaluate
import warnings
warnings.filterwarnings('ignore')

X, y = load_breast_cancer(return_X_y=True, as_frame=True)
X["target"] = y        

X
```

```python
generator = Plugins().get("dummy_sampler")
generator.fit(X, y)

X_gen = generator.generate(1000)

print("Dummy resampler")
evaluate(
    X.drop(columns=["target"]),
    X["target"],
    X_gen.drop(columns=["target"]),
    X_gen["target"],
)
```

```python
generator = Plugins().get("random_noise")
generator.fit(X, y)

X_gen = generator.generate(1000)

print("Random noise")
evaluate(
    X.drop(columns=["target"]),
    X["target"],
    X_gen.drop(columns=["target"]),
    X_gen["target"],
)
```
